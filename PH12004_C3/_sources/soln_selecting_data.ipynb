```python
import pandas as pd

exo_df = pd.read_csv('./data/Exoplanet_catalog_2019.csv')
```

## Solution: Selecting data


Objects with orbital period longer than 250 days:

```python
exo_df[exo_df['orbital_period'] > 250.]
```

Objects with eccentricity below 0.15:

```python
exo_df[exo_df['eccentricity'] < 0.15]
```

Objects with orbital period longer than 250 days **and** eccentricity below 0.15:



```python
exo_df[(exo_df['orbital_period'] > 250.) & (exo_df['eccentricity'] < 0.15)]
```
